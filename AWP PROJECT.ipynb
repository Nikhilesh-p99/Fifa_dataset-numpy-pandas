```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize=(20,5))
sns.countplot(data=df,x="Potential")
plt.grid(True)
plt.show()
#Countplot represent the Potential data
```

```python
df[['Age']].hist(figsize=(7,5),bins=6)
plt.show()
#Histogram represent the age data
```

```python
sns.distplot(df['Age'])
#Distribution plot respresent the age dta
```

```python
plt.figure(figsize=(20,5))
mprice=df["Age"].mean()
meprice=df["Age"].median()
sns.distplot(df["Age"])
plt.axvline(mprice,color='r')
plt.axvline(meprice,color='y')
plt.grid(True)
# Mean is shown in red where as median is shown in yellow
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=df,y="Age")
plt.grid(True)
#Count plot shows the age data
```

```python
sns.pairplot(df[["Skill Moves","Age",]],diag_kind="hist")
plt.show()
#pairplot shows the pairing of skill moves and age data
```

```python
plt.figure(figsize=(23,10))
sns.boxplot(data=df,x='Skill Moves',y='Age')
plt.show()
#boxplot shows skill moves age wise
```
